```python
# 按照用户id进行分类存储到list
def sorted_by_time(dataframe):
    df1_user_id_list=[]
#     根据用户id分类
    user_id_set=set(dataframe.用户标识.tolist())
    for temp_user_id_set in user_id_set:
#         根据用户id 存储到dataframe
        temp_dataframe=dataframe[dataframe.用户标识==temp_user_id_set]
#     根据业务发生时间进行排序
        temp_dataframe=temp_dataframe.sort_values(by='业务发生时间',ascending=True)
        df1_user_id_list.append(temp_dataframe)
#         返回最终结果
    return df1_user_id_list
```

```python
df1_user_id_list=sorted_by_time(df1)
```
